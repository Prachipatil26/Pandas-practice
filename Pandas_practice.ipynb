Time series in Pandas refers to data indexed by date/time.
Pandas provides functions like:

✅ to_datetime()

✅ set_index()

✅ resample()

✅ rolling()

✅ shift()


```python
import pandas as pd

data = {
    "Date": pd.date_range("2024-01-01", periods=7),
    "Sales": [100, 200, 150, 300, 250, 400, 350]
}

df = pd.DataFrame(data)
df.set_index("Date", inplace=True)

print("Original Data:")
print(df)

print("\nWeekly Average:")
print(df.resample("W").mean())

print("\nRolling Average:")
df["Rolling"] = df["Sales"].rolling(3).mean()
print(df)
```

```python
df.index.year
df.index.month
df.index.day
```

```python
#Time Series Shifting (Lag Features)
df["Previous_Day_Sales"] = df["Sales"].shift(1)
print(df)
```

```python
df["Previous_Day_Sales"] = df["Sales"].shift(2)
print(df)
```

```python
#Selecting Data by Date

#select one day
df.loc["2024-01-02"]
```

```python
#select range
df.loc["2024-01-01":"2024-01-03"]
```

Data Visualization in Pandas

Data Visualization means representing data in the form of graphs and charts to understand patterns, trends, and insights easily.

Pandas provides built-in plotting functions using Matplotlib.

```python
import pandas as pd
s1= pd.Series(data=['A','b','A','b','C'])
s2=pd.Series(data=[10,20,30,40,50])
Data1=pd.DataFrame({
    'name': s1,
    'Lemonades':s2
})
Data1['Lemonades'].plot()
```

```python
Data1['Lemonades'].plot(kind='bar')
```
